```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

# Cargar datos
df = pd.read_excel('aerosol.xlsx')
df['Tipo de Aerosol'] = df['Tipo de Aerosol'].str.strip()

# Categorías y colores
categorias = {
    'Continental': ['Continental limpio', 'Continental Promedio', 'Continental Contaminado'],
    'Marítimo': ['Marítimo limpio', 'Marítimo contaminado', 'Marítimo tropical'],
    'Polar': ['Ártico', 'Antártico'],
    'Urbano': ['Urbano'],
    'Desértico': ['Desértico']
}

colores = {
    'Continental': 'tab:green',
    'Marítimo': 'tab:blue',
    'Polar': 'tab:purple',
    'Urbano': 'tab:red',
    'Desértico': 'tab:brown'
}

# Función para asignar categoría
def obtener_categoria(tipo):
    for cat, nombres in categorias.items():
        if tipo in nombres:
            return cat
    return 'Otro'

# Asignar categoría y color
df['Categoría'] = df['Tipo de Aerosol'].apply(obtener_categoria)
df['Color'] = df['Categoría'].map(colores)

# Crear gráfico
fig, ax = plt.subplots(figsize=(11, 6))

# Dibujar puntos
for i in range(len(df)):
    x = df['AOD'][i]
    y = df['ALFA(500-800)'][i]
    color = df['Color'][i]
    ax.scatter(x, y, color=color, marker='s', s=200)

# Etiquetas de cada punto (puedes ajustar libremente)
ax.text(0.009 + 0.01, 1.5 + 0.03, 'Continental limpio', fontsize=12, fontweight='bold')
ax.text(0.09 + 0.01, 1.25 + 0.03, 'Continental Promedio', fontsize=12, fontweight='bold')
ax.text(0.25 + 0.01, 1.25 + 0.03, 'Continental Contaminado', fontsize=12, fontweight='bold')
ax.text(0.573 + 0.01, 1.375 + 0.03, 'Urbano', fontsize=12, fontweight='bold')
ax.text(0.29 + 0.01, 0.125 + 0.03, 'Desértico', fontsize=12, fontweight='bold')
ax.text(0.1 + 0.01, 0.04 + 0.03, 'Marítimo limpio', fontsize=12, fontweight='bold')
ax.text(0.04 + 0.01, 0.43 + 0.03, 'Marítimo contaminado', fontsize=12, fontweight='bold')
ax.text(0.003 + 0.01, 0.15 + 0.03, 'Marítimo tropical', fontsize=12, fontweight='bold')
ax.text(0.005 + 0.01, 0.83 + 0.03, 'Ártico', fontsize=12, fontweight='bold')
ax.text(0.075 + 0.01, 0.68 + 0.03, 'Antártico', fontsize=12, fontweight='bold')

# Etiquetas de ejes
ax.set_xlabel(r'AOD (550 nm)', fontsize=14, fontweight='bold')
ax.set_ylabel(r'AE (500-800 nm)', fontsize=14, fontweight='bold', labelpad=10)

# Ticks principales
ax.tick_params(axis='both', which='major', labelsize=12, width=1.5, length=6)

# Minor ticks
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(axis='both', which='minor', width=1.2, length=4)

# Grueso de los bordes de los ejes
for spine in ax.spines.values():
    spine.set_linewidth(2)

# Rango y ticks personalizados
ax.set_xticks([0.0, 0.2, 0.4, 0.6])
ax.set_yticks([0.0, 0.5, 1.0, 1.5])
plt.yticks(rotation=90)

# Desplazar ligeramente los ticks del eje X
for tick in ax.get_xticklabels():
    tick.set_horizontalalignment('left')
    tick.set_x(tick.get_position()[0] + 0.005)

# Hacer los números de los ejes en negrita
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontweight('bold')

# Límites del gráfico
ax.set_xlim(0, 0.7)
ax.set_ylim(0, 1.7)

# Mostrar y guardar
plt.tight_layout()
plt.savefig('AOD_vs_AE.png', dpi=300)  # Alta resolución
plt.show()
```
